## <font color=red>windows 电脑上代码运行环境配置方法</font>

1. 安装 Anaconda，并配置环境变量，[Anaconda下载地址](https://pan.baidu.com/s/1ocZi0X0r5sfM1iBhiln7Lg)，提取码：0bhk
2. 更换 Anaconda 的镜像源，命令行输入：
    - conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free/
    - conda config --set show_channel_urls yes
3. 删除目录 C:\Users\\<你的用户名> 下配置文件.condarc中的-default

4. 升级 pip，命令行输入： python -m pip install --upgrade pip，若出现 Read timed out ，则使用： python -m pip install --timeout=100 --upgrade pip
5. 安装 TensorFlow 库，命令行输入：pip install tensorflow==1.4.0
6. 安装 PyTorch 库：
    - 下载 torch-0.4.0-cp36-cp36m-win_amd64.whl 到本地，[下载地址](https://pan.baidu.com/s/1ocZi0X0r5sfM1iBhiln7Lg)，提取码：0bhk
    - 在下载的文件目录下打开命令行窗口，输入：pip install torch-0.4.0-cp36-cp36m-win_amd64.whl
6. 安装 Keras 库，不同课件的代码中需要的 Keras 版本不一致，具体到课件中会提示如何安装
7. 安装 MXNet 库，命令行输入：pip install mxnet
8. 安装 PaddlePaddle库：pip install paddlepaddle
8. 安装 requests、cherrypy 、paste、aiohttp 等模块，安装方法：pip install "模块名"
